###    Parsing information from name of files
    Database is stored semi-structural in the general folder in the .wav format
    For the confidental purpose, the actual data not provided in this repository, only pseudo data is used
    Parsed name not in a fixed format, below is an example:
        (Shoot order) (address) (length1) ft (material1) (diameter1) in ... 
        length2 ft (material2) (diameter2) in W@ (location1) B@ (location2) HF LD.wav

```python
import os
import pandas as pd
import numpy as np
import re
```

```python
def getListOfFiles(dirName):
    # create a list of file and sub directories 
    # names in the given directory 
    listOfFile = os.listdir(dirName)
    allFiles = list()
    # Iterate over all the entries
    for entry in listOfFile:
        # Create full path
        fullPath = os.path.join(dirName, entry)
        # If entry is a directory then get the list of files in this directory 
        if os.path.isdir(fullPath):
            allFiles = allFiles + getListOfFiles(fullPath)
        else:
            allFiles.append(fullPath)               
    return allFiles        
 
def main():
    
    dirName = 'C:/Hieu 2018/Study/After grad/Updated data January/Leak Detection pseudo'
       
    # Get the list of all files in directory tree at given path
    listOfFiles = getListOfFiles(dirName)
    
    # Get the list of all files in directory tree at given path
    listOfFiles = list()
    for (dirpath, dirnames, filenames) in os.walk(dirName):
        listOfFiles += [os.path.join(dirpath, file) for file in filenames]
        
    final_list  =[]  
    # Print the files    
    for elem in listOfFiles:
        elem_SPLIT= elem.split('.')
        # only .wav files selected
        if elem_SPLIT[-1]=='wav':
            final_list.append(elem)
        
    return final_list
     
if __name__ == '__main__':
    final_list=main()

# print
for i in final_list:
        elem_SPLIT= i.split('Leak Detection pseudo\\')
        print(elem_SPLIT[1])
```

```python
file_name=[]
year=[]
street=[]
final_list_short=[]
for i in final_list:
        elem_SPLIT= i.split('\\')
        file_name.append(elem_SPLIT[-1])
        year.append(elem_SPLIT[1])
        street.append(elem_SPLIT[2])
        
        elem_SPLIT2= i.split('Leak Detection pseudo')
        final_list_short.append('Leak Detection pseudo'+elem_SPLIT2[1])

df = pd.DataFrame({'directory':final_list_short})
df.insert(1, "Year", year)
df.insert(2, "Street", street)
df.insert(3, "File name", file_name)
```

```python
diameters1=[]
diameters2=[]

for i in file_name: 
    elem_SPLIT=re.split('in|in_|in.', i, flags=re.IGNORECASE)
    print(elem_SPLIT)
    
    elem_SPLIT[0]=elem_SPLIT[0].rstrip()
    SPLIT=elem_SPLIT[0].split(' ')        
    try:
        num1=float(SPLIT[-1])
        diameters1.append(num1)
    except:
        diameters1.append(np.nan)
    
    try:       
        elem_SPLIT[1]=elem_SPLIT[1].rstrip()
        SPLIT=elem_SPLIT[1].split(' ')        
        try:
            num1=float(SPLIT[-1])
            diameters2.append(num1)
        except:
            diameters2.append(np.nan)
    except:
        diameters2.append(np.nan)
df.insert(4, "Diamter 1", diameters1)
df.insert(5, "Diamter 2", diameters2)        
```

```python
length1=[]  
length2=[]      
for i in file_name: 
    elem_SPLIT=re.split('ft|fft|ft.|ft_', i, flags=re.IGNORECASE) 

    elem_SPLIT[0]=elem_SPLIT[0].rstrip()
    SPLIT=elem_SPLIT[0].split(' ')        
    try:
        num1=float(SPLIT[-1])
        length1.append(num1)
    except:
        length1.append(np.nan)        
        
    try:       
        elem_SPLIT[1]=elem_SPLIT[1].rstrip()
        SPLIT=elem_SPLIT[1].split(' ')        
        try:
            num1=float(SPLIT[-1])
            length2.append(num1)
        except:
            length2.append(np.nan)
    except:
        length2.append(np.nan)    

df.insert(6, "Length 1", length1)
df.insert(7, "length 2", length2)
```

```python
material1=[]  
material2=[] 

for i in file_name:
    li_mat_loc=[]
    try:
        DI_loc=i.lower().find(' di ')
        if DI_loc >0 :
            li_mat_loc.append(DI_loc)   
            
        COPPER_loc=i.lower().find(' copper ')
        if COPPER_loc >0:
            li_mat_loc.append(COPPER_loc)
        
        CI_loc=i.lower().find(' ci ')
        if CI_loc >0:
            li_mat_loc.append(CI_loc)    
        
        order=sorted(li_mat_loc)
 
        if order[0]==DI_loc:
            material1.append("DI")
        elif order[0]==CI_loc:
            material1.append("CI")
        elif order[0]==COPPER_loc:
            material1.append("Copper")
        else:
            material1.append(np.nan)
        
        try:
            if order[1]==-1:
                material2.append(np.nan)
            elif order[1]==DI_loc:
                material2.append("DI")
            elif order[1]==CI_loc:
                material2.append("CI")
            elif order[1]==COPPER_loc:
                material2.append("Cooper")
        except:
            material2.append(np.nan)
    except:
        material1.append(np.nan)
        material2.append(np.nan)
            
df.insert(8, "Material 1", material1)
df.insert(9, "Material 2", material2)
```

```python
first_at=[]  
second_at=[]      
for i in file_name: #[file_name[0],file_name[1]]:
#    elem_SPLIT= i.split('ft')
#    try:
    elem_SPLIT=re.split('W @|W@|B @|B@|HF LD|HFLD', i, flags=re.IGNORECASE)
    if len(elem_SPLIT)>2:
        first_at.append(elem_SPLIT[1])
        second_at.append(elem_SPLIT[2])
    else:   
        first_at.append(np.nan)
        second_at.append(np.nan)
df.insert(10, "First station at", first_at)
df.insert(11, "Second station at", second_at)    
```

```python
color_first=[] 
color_second=[] 
for i in file_name:
    white_loc=max((i.lower().find('w@'),i.lower().find('w @')))
    blue_loc=max((i.lower().find('b@'),i.lower().find('b @')))
    
    if white_loc==-1 or blue_loc==-1:
        color_first.append(np.nan)
        color_second.append(np.nan)
        continue
    
    
   
    if white_loc<blue_loc:
        color_first.append('W')
        color_second.append('B')
    elif white_loc>blue_loc:
        color_first.append('B')
        color_second.append('W')

df.insert(12, "First station color", color_first)
df.insert(13, "Second station color", color_second) 
```
